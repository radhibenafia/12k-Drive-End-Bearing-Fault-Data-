```python
import numpy as np
import matplotlib.pyplot as plt
from pyts.image import GramianAngularField

# Fixer la graine aléatoire pour la reproductibilité
np.random.seed(42)

# Créer une série temporelle linéaire avec un peu de bruit
n_points = 100
temps = np.arange(n_points)
serie_temporelle = 0.5 * temps + 10 + np.random.normal(scale=5, size=n_points)

# Normaliser la série temporelle entre 0 et 1
serie_temporelle_normalisee = (serie_temporelle - np.min(serie_temporelle)) / (np.max(serie_temporelle) - np.min(serie_temporelle))

# Utiliser Gramian Angular Field pour transformer la série temporelle
gaf = GramianAngularField(image_size=10, method='difference')
serie_temporelle_gaf = gaf.fit_transform(serie_temporelle_normalisee.reshape(1, -1))

# Afficher la série temporelle
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(temps, serie_temporelle, label='Série Temporelle')
plt.xlabel('Temps')
plt.ylabel('Valeurs')
plt.title('Exemple de Série Temporelle')
plt.legend()

# Afficher l'image Gramian Angular Field
plt.subplot(1, 2, 2)
plt.imshow(serie_temporelle_gaf[0], cmap='rainbow', origin='lower')
plt.title('Gramian Angular Field')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyts.image import RecurrencePlot

# Fixer la graine aléatoire pour la reproductibilité
np.random.seed(42)

# Créer une série temporelle linéaire avec un peu de bruit
n_points = 100
temps = np.arange(n_points)
serie_temporelle = 0.5 * temps + 10 + np.random.normal(scale=5, size=n_points)

# Normaliser la série temporelle entre 0 et 1
serie_temporelle_normalisee = (serie_temporelle - np.min(serie_temporelle)) / (np.max(serie_temporelle) - np.min(serie_temporelle))

# Utiliser Recurrence Plot pour transformer la série temporelle
rp = RecurrencePlot(dimension=1, percentage=30)
serie_temporelle_rp = rp.fit_transform(serie_temporelle_normalisee.reshape(1, -1))

# Afficher la série temporelle
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(temps, serie_temporelle, label='Série Temporelle')
plt.xlabel('Temps')
plt.ylabel('Valeurs')
plt.title('Exemple de Série Temporelle')
plt.legend()

# Afficher le Recurrence Plot
plt.subplot(1, 2, 2)
plt.imshow(serie_temporelle_rp[0], cmap='binary', origin='lower')
plt.title('Recurrence Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyts.image import RecurrencePlot
from sklearn.preprocessing import MinMaxScaler

# Générer une série temporelle sinusoïdale
n_points = 100
temps = np.linspace(0, 4 * np.pi, n_points)
serie_temporelle_angulaire = np.sin(temps)

# Introduire une coupure dans la série temporelle
coupure_index = 60
serie_temporelle_angulaire[coupure_index:] = 0

# Normaliser la série temporelle entre 0 et 1
scaler = MinMaxScaler(feature_range=(0, 1))
serie_temporelle_angulaire_normalisee = scaler.fit_transform(serie_temporelle_angulaire.reshape(-1, 1)).flatten()

# Utiliser Recurrence Plot pour transformer la série temporelle
rp = RecurrencePlot(dimension=1, percentage=30)
serie_temporelle_rp = rp.fit_transform(serie_temporelle_angulaire_normalisee.reshape(1, -1))

# Afficher la série temporelle
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(temps, serie_temporelle_angulaire, label='Série Temporelle Angulaire avec Coupure')
plt.xlabel('Temps')
plt.ylabel('Valeurs')
plt.title('Exemple de Série Temporelle Angulaire avec Coupure')
plt.legend()

# Afficher le Recurrence Plot
plt.subplot(1, 2, 2)
plt.imshow(serie_temporelle_rp[0], cmap='binary', origin='lower')
plt.title('Recurrence Plot après la Coupure')
plt.show()


```
